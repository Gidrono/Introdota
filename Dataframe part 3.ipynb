# filter by multiple conditions (AND/OR)

## AND

```python
# first way
bond[(bond["Bond actor"] == "Sean Connery") and bond["Academy Award nominations"] ==1]# elaborate on this value error https://stackoverflow.com/questions/21415661/logical-operators-for-boolean-indexing-in-pandas
```

```python
bond[(bond["Bond actor"] == "Sean Connery") & (bond["Academy Award nominations"] ==1)]
```

```python
# first way is kind of messy and unclear so lets move to the second way:
mask1 = bond["Bond actor"] == "Sean Connery" # return boolean values
mask2 = bond["Academy Award nominations"] ==1 # Series of Boolean Values
bond[mask1 & mask2] # dataframe with actual values
```

### OR

```python
bond[(bond["Bond actor"] == "Sean Connery") | (bond["Year"] == 2008)]
mask1 = bond["Bond actor"] == "Sean Connery"
mask2 = bond["Year"] == 2008
bond[mask1|mask2]
```

```python
#combine
mask1 = bond["Bond actor"] == "Sean Connery"
mask2 = bond["Year"] > 1962
mask3 = bond["Academy Award wins"] >0
bond[(mask1&mask2) | mask3]
```

# .isnull() / isna()/notnull()/notna() methods

```python
nba = pd.read_csv("nbasalariespoints.csv", usecols = ["Player", "Age", "SALARY"])
nba.rename(columns = {"SALARY": "Salary"}, inplace = True)
nba.head()
```

```python
nba[nba["Age"].isna()].head() # isnull()
```

```python
nba[nba["Age"].notnull()].tail() #notna()
```

# Dealing with duplicates

```python
nba.sort_values("Age", inplace=True)
nba.head()
```

```python
# identify duplicates in a column/Series:
nba.Age.duplicated().head()
```

```python
# as you noticed the first value is NOT considered as False. that because inside the duplicated method there is a parameter keep = "first" which mean Mark duplicates as True except for the first occurrence. LETS CHANGE IT
nba.Age.duplicated(keep = "last").head(10)
#nba[nba.Age.duplicated(keep = "last")].head(10)
```

```python
nba.Age.duplicated(keep = False).head(10) # Mark all duplicates as True
```

```python
# another way:
nba[nba.Age.duplicated() == True].head()
```

```python
#on a dataframe its basically the same:
nba[nba.duplicated(subset = "Age")].head()
```

```python
nba[nba.duplicated(subset = ["Age", "Salary"])].head()
# risky if you have null values. 
nba[nba.duplicated(subset = ["Age", "Salary"], keep=False)].head()
```

```python
# extract only not_duplicated:
#first way:
nba[nba.duplicated(subset = "Age") == False].head() 
# or use tilda ~ 
nba[~nba.Age.duplicated()].head()
```

## drop_duplicates

```python
nba.drop_duplicates(subset = 'Age') # and dont forget the inplace=True if ypu want to change it on the original dataframe
```

```python
nba.drop_duplicates(subset = ['Age', "Salary"], inplace = True)
nba.shape
```
