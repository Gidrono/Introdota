# filter with .where method

```python
bond.where(bond["Bond actor"]=="Sean Connery", axis=1).dropna(how = "all")
```

```python
mask = bond["Bond actor"]=="Roger Moore"
bond.where(mask).dropna(how="all")
```

```python
mask2 = bond["Actual box office"] >100
bond.where(mask&mask2).dropna(how="all")
```

```python
# create new column using np.where
data = {'name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
        'age': [42, 52, 36, 24, 73], 
        'preTestScore': [4, 24, 31, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70]}
df = pd.DataFrame(data, columns = ['name', 'age', 'preTestScore', 'postTestScore'])
df
```

```python
# Create a new column called df.elderly where the value is yes
# if df.age is greater than 50 and no if not
df['elderly'] = np.where(df['age']>=50, 'yes', 'no')
df
```
