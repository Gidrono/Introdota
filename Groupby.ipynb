A groupby operation involves some combination of splitting the object, applying a function, and combining the results.
This can be used to group large amounts of data and compute operations on these groups.

```python
fortune.groupby("Sector")
```

```python
print (len(fortune.groupby("Sector"))) # how many section? 
fortune.groupby("Sector").size() # same as 
#fortune["Sector"].value_counts()
```

```python
# filter out all companies based on a condition
fortune.groupby(['Sector', 'Industry']).filter(lambda x: 
             x["Employees"].sum()>2000000)
```

```python
#first let create a data frame
df=pd.DataFrame(columns=fortune.columns)
df
```

```python
sectors=fortune.groupby("Sector")
for sector,data in sectors: # loop through all groups to get the company with the highest revenue for that grouping
    highest_revenue_compnay_in_group=data.nlargest(1, "Revenue")
    df = df.append(highest_revenue_compnay_in_group)
```

```python
df = pd.DataFrame(columns=fortune.columns)
for city, data in fortune.groupby("Location"): # BEST PERFORMING COMPANY IN every city
    highest_revenue_in_city=data.nlargest(1, "Revenue")
    df=df.append(highest_revenue_in_city)
df.head(3)
```
