```python
import pandas as pd
```

```python
df = pd.read_csv("James Bond 007 Movie Data.csv")
df.head()
```

# string methods from python works also in pandas 
some examples

```python
df["Bond actor"].str.capitalize().unique()
```

```python
df["Bond actor"].str.find("Connery") # remember that find return the index and -1 if not exist
df["Bond actor"].str.len().head() # return numbers
df["Bond actor"].str.split().head()# return list
df["Bond actor"].str.isalpha().head() #return boolean values. also why False?
```

```python
df["new_name"] = df["Bond actor"].apply(lambda x: "".join(x.split()))
df["new_name"].str.isalpha().head()
```

# methods using regex build in
some example

```python
df["Bond actor"].str.replace("Sean", "Sir Sean").head() # re.sub()
df["Bond actor"].str.replace("sean", "Sir Sean", case=True).head() # if case = True (default) it mean case sensitive)
df["Bond actor"].str.replace("sean", "Sir Sean", case=False, regex=True).head()# IF REGEX = False treats the pattern as a literal string 
```

```python
df[df["Bond actor"].str.contains("S")].tail()# return boolean unless you "wrapped" it again with df
```

```python
df[df["Bond actor"].str.contains("S|D", case=None)].tail()# (s or d)
```

```python
df[df["Bond actor"].str.contains("S|D", case=None)]#["Bond actor"]
```

```python
df[df["Bond actor"].str.contains("^[^S].*[^g]$")] #find all names that not starting with "S" and not ending with a "g", making use of the start-of-string (^) and end-of-string ($) 
```
